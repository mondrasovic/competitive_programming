```python
def n_choose_r_mod_p_pascal(n: int, r: int, m: int) -> int:
    if r == 0 or r == n:
        return 1
    if r > n:
        return 0

    n_choose_r = [0 for _ in range(r + 1)]
    n_choose_r[0] = 1

    for i in range(1, n+1):
        for j in range(min(r, i), 0, -1):
            n_choose_r[j] = (n_choose_r[j] + n_choose_r[j - 1]) % m

    return n_choose_r[r]

def n_choose_r_mod_p_lucas(n: int, r: int, m: int) -> int:
    if r == 0:
        return 1
    if r > n:
        return 0
     
    return (
        n_choose_r_mod_p_lucas(n // m, r // m, m) * n_choose_r_mod_p_pascal(n % m, r % m, m)
    ) % m

def inv(a: int, m: int) -> int:
    m0 = m  
    x0 = 0
    x1 = 1
  
    if m == 1:
        return 0
  
    while a > 1:
        q = a // m  
  
        t = m  

        m = a % m  
        a = t  
  
        t = x0  
  
        x0 = x1 - q * x0  
  
        x1 = t  
      
    if x1 < 0:
        x1 = x1 + m0  
  
    return x1  
  
def chinese_remainder(nums: list[int], rems: list[int]):
    prod = 1
    for i in range(0, len(nums)):
        prod = prod * nums[i]  
  
    result = 0
  
    for i in range(0, len(nums)):
        pp = prod // nums[i]  
        result = result + rems[i] * inv(pp, nums[i]) * pp  
      
    return result % prod 

N = 10 ** 18
K = 10 ** 9

result = 0

for i, p in enumerate(primes):
    res_p = n_choose_r_mod_p_lucas(N, K, p)

    for j in range(i + 1, len(primes)):
        q = primes[j]
        res_q = n_choose_r_mod_p_lucas(N, K, q)

        for k in range(j + 1, len(primes)):
            r = primes[k]
            res_r = n_choose_r_mod_p_lucas(N, K, q)
            result += chinese_remainder([p, q, r], [res_p, res_q, res_r])
```
